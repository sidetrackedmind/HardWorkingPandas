```python
import cs
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
%matplotlib inline
```

```python
com_ids = cs.make_lookup_set('data/EXTR_CommBldg.csv')
res_ids = cs.make_lookup_set('data/EXTR_ResBldg.csv')
ucomm_ids = set([i for i in com_ids if i not in res_ids])
ures_ids = set([i for i in res_ids if i not in com_ids])
```

```python
com_prices = cs.get_rel_prices('data/EXTR_RPSale.csv', ucomm_ids, 10000)
res_prices = cs.get_rel_prices('data/EXTR_RPSale.csv', ures_ids, 10000)
```

```python
com_summary = cs.get_price_summary(com_prices, np.median)
res_summary = cs.get_price_summary(res_prices, np.median)
```

```python
# min year = 1980
com_summary = [i for i in com_summary if i[0] > 1980]
res_summary = [i for i in res_summary if i[0] > 1980]
comm_x = [i[0] for i in com_summary]
comm_y = [i[1] for i in com_summary]
res_x = [i[0] for i in res_summary]
res_y = [i[1] for i in res_summary]
```

```python
plt.plot(comm_x, comm_y, label = 'Commerical')
plt.plot(res_x, res_y, label = 'Residential')
plt.legend()
plt.title('Commercial vs Residential Real Estate Sale Prices in Seattle 1980 - 2017')
plt.show()
```
